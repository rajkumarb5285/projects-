```python
step:5- Missing value tretments
```

```python
#choose the method of imuting missing value
#like mean, median,mode or KNNImputer
```

```python
# List of columns to process
columns_to_fill = ["BMI", "Polio", "Income composition of resources"]

# Iterate through each column and fill missing values with the median
for col in columns_to_fill:
    if col in df.columns:  # Check if column exists in the DataFrame
        df[col] = df[col].fillna(df[col].median())
```

```python
from sklearn.impute import KNNImputer
impute=KNNImputer()
```

```python
for i in df.select_dtypes(include="number").columns:
     df[i]=impute.fit_transform(df[[i]]) 
```

```python
step6:-Outlier treatments
```

```python
#decide wheather to do outlier treatment or not , if do how?
```

```python
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw     
```

```python
wisker(df['GDP']
```

```python
for i in ['GDP','Total expenditure',' thinness  1-19 years',' thinness 5-9 years',]:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
```
